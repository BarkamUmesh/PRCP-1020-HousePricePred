```python
# IQR  method 
Q1 = df[num_cols].quantile(0.25)
Q2 = df[num_cols].quantile(0.50)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3-Q1
df = df[~((df[num_cols] < (Q1 - 1.5*IQR)) | 
          (df[num_cols] > (Q3 + 1.5*IQR))).any(axis=1)]
```

```python
Q1_new = df[num_cols].quantile(0.25)
Q3_new = df[num_cols].quantile(0.75)
IQR_new = Q3_new - Q1_new

outliers_remaining = ((df[num_cols] < (Q1_new - 1.5*IQR_new)) | 
                      (df[num_cols] > (Q3_new + 1.5*IQR_new))).sum()

outliers_remaining
```

```python
num_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy="median")),
    ("scaler",OneHotEncoder(handle_unknown="ignore"))
])
cat_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("scaler",OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num",num_pipeline,num_cols),
    ("cat",cat_pipeline,cat_cols)
])
```

##### Why median for Numerical ?
- Robust to outliers
- House prices often have extreme values
- Mean gets distorted
##### Business logic:
- Median represents a typical house better than mean.

#####  Why most frequent(mode) for Categorical?
- Keeps real category
- Doesn’t introduce artificial labels
-Better than filling "None" manually
##### Example:
- If most houses have GarageType = Attached, missing likely means same.


We use SimpleImputer to handle missing values in a statistically meaningful way. Median is chosen for numeric features because it is robust to outliers, and most-frequent is chosen for categorical features to preserve common patterns.

We use OneHotEncoder instead of Label or Ordinal Encoding because most categorical variables in house price data are nominal and have no inherent order. OneHotEncoding avoids introducing false ranking and allows models to learn independent effects of each category.

Using all preprocessing steps inside a pipeline makes the solution scalable, reproducible, and production-ready.

```python
param_grid = {
    "n_estimators":[150,200] ,
    "learning_rate":[0.03,0.05],
    "max_depth": [2,3],
    "subsample":[0.7,0.8],
    "min_samples_leaf":[5,10]
}

gbr = GradientBoostingRegressor(random_state=42)

# GridSearch
grid_search = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

# Fit GridSearch
grid_search.fit(X_train, y_train)

# Best model
best_gb = grid_search.best_estimator_

print("Best Parameters:", grid_search.best_params_)
```

#### Create a report which should include the challenges faced on data and what techniques were used with proper reasoning

1. Missing Values
- Challenge
- Several numerical and categorical features contained missing values, which could lead to biased models or errors during training.
- Technique Used
- Numerical features: Median imputation
- Categorical features: Mode imputation
- Reasoning
- The median is less sensitive to extreme values than the mean and preserves the central tendency of skewed numerical distributions. Mode imputation maintains the most frequent category and helps preserve the original distribution of categorical features.

2. Outliers
- Challenge
- Some features such as lot area, living area, and sale price had extreme values that could distort model learning.
- Technique Used
- Interquartile Range (IQR) method was applied to detect and cap outliers.
- Reasoning
- The IQR method is robust for skewed data and does not assume normal distribution, making it suitable for real estate data where values are often unevenly distributed.

3. Categorical Variables
- Challenge
- The dataset contains many categorical variables (e.g., neighborhood, house style, roof type), which machine learning models cannot process directly.
- Technique Used
- One-Hot Encoding was applied to convert categorical variables into numerical format.
- Reasoning
- One-Hot Encoding prevents models from assuming any ordinal relationship between categories and works effectively with tree-based models such as Gradient Boosting and XGBoost.

 4. Feature Scaling
- Challenge
- Numerical features were on different scales, which can affect the performance of distance-based and linear models.
- Technique Used
- StandardScaler was used to standardize numerical features.
- Reasoning
- Scaling ensures that all features contribute equally to model training and improves convergence for linear and regularized regression models.

5. High Dimensionality
- Challenge
- After encoding, the number of features increased significantly, increasing model complexity.
- Technique Used
- Regularization (Ridge and Lasso regression) and tree-based models were applied.
- Reasoning
- Regularization reduces overfitting by penalizing large coefficients, while tree-based models naturally handle large feature spaces.

6. Multicollinearity
Challenge
- Some numerical features were highly correlated, which can affect linear model stability.
- Technique Used
- Correlation analysis and Ridge regression.
- Reasoning
- Ridge regression reduces the impact of correlated features by shrinking coefficients and improving generalization.

7. Skewed Target Variable
Challenge
- House prices showed right-skewed distribution.
Technique Used
- Log transformation of target variable (optional step).
- Reasoning
- Log transformation makes the distribution more normal-like and improves regression model performance.
